```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data1 = pd.read_csv("data/output/output1.txt",sep="\t")
```

```python
plt.figure(figsize=(16,8))
plt.plot(data1["px"],data1["py"],label="Predicted")
plt.plot(data1["px_true"],data1["py_true"],label="Ground Truth")
plt.scatter(data1["px_measured"],data1["py_measured"],label="Observation")
plt.legend(loc="best")

plt.xlabel("X",fontsize=20)
plt.ylabel("Y",fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

```python
plt.figure(figsize=(16,8))
plt.plot(data1["NIS"])
plt.hlines(7.815,xmin=0,xmax=len(data1["NIS"]),lw=3)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("k",fontsize=40)
plt.ylabel(r"$\epsilon$",fontsize=40)

plt.xlim([0,len(data1["NIS"])])
```
